```python
import pandas as pd 
import openpyxl
```

```python
#cargar el archivo  de excel es un dataframe
df = pd.read_excel('data\Catalogo1960_2021.xlsx')
```

```python
# Mostrar las primeras filas del DataFrame
df.head()
```

```python
# Obtener informaci√≥n general del DataFrame
df.info()
```

```python
# Ver estad√≠sticas descriptivas de las columnas num√©ricas
df.describe()
```

```python
# Eliminar columnas innecesarias (si las hay)
df = df.drop(['ID', 'FECHA_CORTE'], axis=1)
```

```python
# Convertir las columnas de fecha y hora a tipo datetime
df['FECHA_UTC'] = pd.to_datetime(df['FECHA_UTC'], format='%Y%m%d')
df['HORA_UTC'] = pd.to_datetime(df['HORA_UTC'].astype(str).str.zfill(6), format='%H%M%S').dt.time
```

```python
# Lidiar con los valores faltantes (si los hay)
# puedes rellenar los valores faltantes en la columna 'PROFUNDIDAD' con el valor promedio
df['PROFUNDIDAD'].fillna(df['PROFUNDIDAD'].mean(), inplace=True)
```

```python
# Asegurarse de que todas las columnas tengan los tipos de datos correctos
df['MAGNITUD'] = df['MAGNITUD'].astype(float)
```

```python
#vemos nuestro dataframe ya limpio, y puesto los tipos de datos para el analisis
df.head()
```

```python
#Guardar el DataFrame limpio a csv
df.to_csv('data\Catalogo1960_2021_limpiado.csv', index=False)
```
